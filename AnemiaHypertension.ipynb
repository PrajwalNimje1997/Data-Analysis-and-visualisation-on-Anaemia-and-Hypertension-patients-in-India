```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import geopandas as gpd
import seaborn as sns
```

```python
df1 = pd.read_csv('/Users/prajwalite/Downloads/TRJ_3.3 - Table 1.csv')
df1.head()
```

```python
# Attribute distribution
df2 = df1['STATES']                    #index
df3 = df1['AnaemicMigrants']           #attribute AnaemicMigrants
df4 = df1['AnaemicNonMigrant']         #attribute AnaemicNonMigrants
df5 = df1['HTMigrants']                #attribute HTMigrants
df6 = df1['HTNon-Migrant']             #attribute HTNon-Migrant
```

# Comparison of Anaemia and HT for all the states and India.

```python
AnaemicMigrantsValue = np.array(df3)    #array of df3
AnaemicNonMigrantValue = np.array(df4)  #arrray of df4
HTMigrantsValue = np.array(df5)         #array of df5
HTNonMigrantValue = np.array(df6)       #array of df6
HTNonMigrantValue[:2]
```

```python
anaemicCombined = pd.DataFrame({'migrant': AnaemicMigrantsValue,'non-migrant': AnaemicNonMigrantValue}, index=df2 )#, figsize=(12,10))
anaemicCombined.plot.barh(figsize=(16,12))
plt.title('Percentage of migrant and non migrant women having Anaemia')
plt.xlabel("Percentage")
plt.show()
```

# Interpretation:
From the figure, it is found that the chances of getting diagnosed with Anaemia is more in non-migrant women as compared to migrant women in all the states, in all the regions of the country.

```python
HTCombined = pd.DataFrame({'migrant': HTMigrantsValue,'non-migrant': HTNonMigrantValue}, index=df2 )#, figsize=(12,10))
HTCombined.plot.barh(figsize=(16,12))
plt.title('Percentage of migrant and non migrant women having Hypertension')
plt.xlabel("Percentage")
plt.show()
```

# Interpretation:
It can be seen that the number of women suffering with Hypertension is more in non-migrant women as compared to migrant women. But there are some places where the number of women suffering with Hypertenion is more in migrants than non-migrants. In this place number of migrants suffering with Hypertension is much more that non-migrants. In short it can be concluded that the short term migration is proving beneficial for the women. It is interpreted that the migration is proving good for releasing hypertension among women.

# Representation of intensity of Anaemia and Hypertension among migrant and non-migrant women in India.

```python
shp_gdf = gpd.read_file('/Users/prajwalite/Downloads/india_administrative_state_boundary/india_administrative_state_boundary.shp')
shp_gdf['st_nm'][12] = 'Jammu And Kashmir'
shp_gdf['st_nm'][7] = 'Daman and Diu'
shp_gdf['st_nm'][6] = 'Dadara Nagar Havel'
shp_gdf['st_nm'][23] = 'Delhi'
```

```python
merged = shp_gdf.set_index('st_nm').join(df1.set_index('STATES'))
merged.head()
```

```python
# Intensity of Migrant women having Anaemia
fig, ax = plt.subplots(1, figsize=(12, 12))
ax.axis('off')
ax.set_title('Intensity of migrant women having Anaemia',fontdict={'fontsize': '15', 'fontweight' : '3'})
fig = merged.plot(column='AnaemicMigrants', cmap='seismic', linewidth=0.5, ax=ax, edgecolor='0.2',legend=True)
```

# Interpretation:
It is found that the migrant women in the North and North East are getting more affected than other parts of the country. Women in Bihar, West Bengal, Orissa have high chances of getting affected with Anaemia.

```python
# Intensity of non-migrant women having Anaemia
fig, ax = plt.subplots(1, figsize=(12, 12))
ax.axis('off')
ax.set_title('Intensity of non-migrant women having Anaemia',fontdict={'fontsize': '15', 'fontweight' : '3'})
fig = merged.plot(column='AnaemicNonMigrant', cmap='seismic', linewidth=0.5, ax=ax, edgecolor='0.2',legend=True)
```

# Interpretation:
It is found that the non-migrant women in the East, North and North East, Central,West are getting more affected than other parts of the country. The chances of affecting with Anaemia is high in almost all the parts of the country. Only few states like Kerala and karnataka have low rate of Anaemia among non-migrant women. 

```python
# Intensity of migrant women having Hypertension
fig, ax = plt.subplots(1, figsize=(12, 12))
ax.axis('off')
ax.set_title('Intensity of migrant women having Hypertension',fontdict={'fontsize': '15', 'fontweight' : '3'})
fig = merged.plot(column='HTMigrants', cmap='seismic', linewidth=0.5, ax=ax, edgecolor='0.2',legend=True)
```

# Interpretation:
It is found that the migrant women suffered with Hypertension is less in almost all the parts of the country. The rate is average in the Norther part like Jammu and Kashmir, Haryana, Punjab.

```python
# Intensity of non-migrant women having Hypertension
fig, ax = plt.subplots(1, figsize=(12, 12))
ax.axis('off')
ax.set_title('Intensity of non-migrant women having Hypertension',fontdict={'fontsize': '15', 'fontweight' : '3'})
fig = merged.plot(column='HTNon-Migrant', cmap='seismic', linewidth=0.5, ax=ax, edgecolor='0.2',legend=True)
```

# Interpretation:
It is found that the non-migrant women suffered with Hypertension is less in almost all the parts of the country. The rate is average in the Norther part like Jammu and Kashmir, Haryana, Punjab and the Eastern part in Assam and Nagaland.
